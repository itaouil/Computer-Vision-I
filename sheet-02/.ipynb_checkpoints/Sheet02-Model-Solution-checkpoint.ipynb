## Blurring Using FFT

First we need to pad the kernel to become the same size as the image.

```python
img_h, img_w = image.shape
krn_h, krn_w = kernel_2d.shape
plt.matshow(kernel_2d)
```

```python
pad_up = img_h // 2 - krn_h // 2
pad_down = img_h // 2 - krn_h // 2 - 1
pad_left = img_w // 2 - krn_w // 2
pad_right = img_w // 2 - krn_w // 2 - 1

kernel_2d_padded = np.pad(kernel_2d, pad_width=((pad_up, pad_down), (pad_left, pad_right)), mode="constant", constant_values=0)
plt.matshow(kernel_2d_padded)
```

Now we need to calculate the FFT of the image and the padded kernel

```python
img_freq = np.fft.fft2(image)
kernel_freq = np.fft.fft2(kernel_2d_padded)
```

Now multiplying element-wise and taking the inverse

```python
fft_result_freq = img_freq * kernel_freq
fft_result = np.fft.ifft2(fft_result_freq)
plt.imshow(np.abs(fft_result), cmap="gray")
```

There is a shift!

### Why is there a shift?

Let's look at the frequency response of the gaussian kernel. We know that the Fourier transform of a Gaussian signal is a Gaussian response.

```python
kernel_freq = np.fft.fft2(kernel_2d_padded)
plt.imshow(np.abs(kernel_freq))
```

But this is not Gaussian. This is a shifted Gaussian.

So Let's fix it first.

```python
kernel_freq = np.fft.fft2(kernel_2d_padded)
kernel_freq = np.fft.fftshift(kernel_freq)
plt.imshow(np.abs(kernel_freq))
```

### Shift Fix

Now let's use this shift to fix the issue.

```python
img_freq = np.fft.fft2(image)
kernel_freq = np.fft.fft2(kernel_2d_padded)
fft_result_freq = img_freq * kernel_freq
fft_result = np.fft.ifft2(fft_result_freq)

fft_result = np.fft.fftshift(fft_result)

fft_result = np.abs(fft_result)
plt.imshow(fft_result, cmap="gray")
```

## Are the results the same?

```python
_ = plt.hist(fft_result.reshape(-1))
```

```python
_ = plt.hist(conv_result.reshape(-1))
```

The distributions are very similar. Also the average absolute difference is very small. So they are basically the same.

## Normalized Cross Correlation Implementation

```python
def normalized_cross_correlation(image, template):
    image_h, image_w = image.shape
    template_h, template_w = template.shape

    result = np.zeros(
        (image_h - template_h + 1, image_w - template_w + 1), dtype=np.float32
    )
    
    template_mean = np.mean(template)
    
    for i in range(result.shape[0]):
        for j in range(result.shape[1]):
            patch = image[i : i + template_h, j : j + template_w]
            patch_mean = np.mean(patch)

            norm = np.sqrt(
                np.sum(np.power(template - template_mean, 2))
                * np.sum(np.power(patch - patch_mean, 2))
            )
            result[i, j] = (
                np.sum((template - template_mean) * (patch - patch_mean)) / norm
            )
    
    return result
```

## Template Matching Using Cross Correlation

```python
%time result_ncc = normalized_cross_correlation(image, template)
```

Let's visualize and see what is the response.

```python
plt.matshow(result_ncc, cmap="gray")
plt.colorbar()
```

## Visualization

```python
def draw_rectangle_at_matches(image, template_h, template_w, matches):
    matches = zip(matches[0], matches[1])
    for match in matches:
        cv2.rectangle(
            image,
            (match[1], match[0]),
            (match[1] + template_w, match[0] + template_h),
            0,
        )
    return image
```

# Task 3 - Gaussian Pyramid

```python
image = cv2.imread("data/traffic.jpg", cv2.IMREAD_GRAYSCALE)
template = cv2.imread("data/traffic-template.png", cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap="gray")
plt.figure()
plt.imshow(template, cmap="gray")
```

```python
image = np.float32(image) / 255.0
template = np.float32(template) / 255.0
```

## Create 4 level Gaussian Pyramid

```python
def build_gaussian_pyramid(image, num_levels):
    sigma = 1.0
    result = []
    result.append(image)
    for i in range(num_levels):
        blurred_image = cv2.GaussianBlur(image, ksize=(0, 0), sigmaX=sigma)
        height, width = blurred_image.shape
        image = blurred_image[0::2, 0::2] # sampling
        result.append(image)

    return result
```

```python
my_pyramid = build_gaussian_pyramid(image, num_levels=4)
my_pyramid_template = build_gaussian_pyramid(template, num_levels=4)
```

### Visualizing the Pyramid

```python
for p in my_pyramid:
    plt.figure()
    plt.imshow(p, cmap="gray")
```

## Create Gaussian Pyramid using `cv2.PyrDown`

```python
def build_gaussian_pyramid_opencv(image, num_levels):
    result = []
    result.append(image)
    for i in range(num_levels):
        image = cv2.pyrDown(image)
        result.append(image)
    return result
```

```python
cv_pyramid = build_gaussian_pyramid_opencv(image, num_levels=4)
```

## Fast Pyramid Template Matching

```python
def template_matching_multiple_scales(pyramid_image, pyramid_template, show=False):
    threshold = 0.7
    window_size = 5
    
    # reversing the pyramids
    pyramid_image = pyramid_image[::-1]
    pyramid_template = pyramid_template[::-1]
    
    similarity_map = normalized_cross_correlation(pyramid_image[0], pyramid_template[0])
    
    if show:
        plt.figure()
        plt.imshow(similarity_map, cmap="gray")
        plt.colorbar()
    
    for i in range(1, len(pyramid_image)):
        matches = np.where(similarity_map >= threshold)
        matches = zip(matches[0], matches[1])
        
        similarity_map = np.zeros_like(pyramid_image[i])

        image_h, image_w = pyramid_image[i].shape
        template_h, template_w = pyramid_template[i].shape
        
        
        for match in matches:
            match = [x * 2 for x in match]  # match location in the higher scale
            
            # make sure it doesn't go over the limit
            patch_top_left = max(match[0] - window_size, 0), max(match[1] - window_size, 0)
            
            match[0] += template_h
            match[1] += template_w
            
            # make sure it doesn't go over the limit
            patch_bot_right = min(match[0] + window_size, image_h), min(match[1] + window_size, image_w)
            
            patch = pyramid_image[i][
                patch_top_left[0]:patch_bot_right[0],
                patch_top_left[1]:patch_bot_right[1],
            ]
            
            # compute similarity at patch level
            similarity = normalized_cross_correlation(patch, pyramid_template[i])
            similarity_final = np.zeros_like(patch)
            
            similarity_final[:similarity.shape[0], :similarity.shape[1]] = similarity

            similarity_map_patch = similarity_map[
                patch_top_left[0]:patch_bot_right[0],
                patch_top_left[1]:patch_bot_right[1],
            ]
            # find where the similarity is increated compared to previous matches
            mask = np.float32(similarity_final > similarity_map_patch)
            
            # increase the similarity match on the mask area.
            similarity_map[
                patch_top_left[0]:patch_bot_right[0],
                patch_top_left[1]:patch_bot_right[1],
            ] = (mask * similarity_final + (1 - mask) * similarity_map_patch)
        
        if show:
            plt.figure()
            plt.imshow(similarity_map, cmap="gray")
            plt.colorbar()
    
    return similarity_map
```

```python
start = time.time()
similarity_map = template_matching_multiple_scales(
    my_pyramid, my_pyramid_template
)
elapsed = time.time()
print("Task3::Time_Matching_with_pyramid:{0}".format(elapsed - start))
```

## Weights of derivative in x and y

Here is the formula for 1D gaussian

![d of gaussian](http://campar.in.tum.de/twiki/pub/Chair/HaukeHeibelGaussianDerivatives/_MathModePlugin_dd52147e95f0297ca96e5b8c6b281758.png)

```python
def get_exponent(x, y, sigma):
    return -1 * (x * x + y * y) / (2 * sigma)

def get_derivative_of_gaussian_kernel(size, sigma):
    assert size > 0 and size % 2 == 1 and sigma > 0

    kernel_x = np.zeros((size, size))
    kernel_y = np.zeros((size, size))

    size_half = size // 2

    for i in range(size):
        y = i - size_half
        for j in range(size):
            x = j - size_half
            kernel_x[i, j] = (
                -1
                * (x / (2 * np.pi * sigma * sigma))
                * np.exp(get_exponent(x, y, sigma))
            )
            kernel_y[i, j] = (
                -1
                * (y / (2 * np.pi * sigma * sigma))
                * np.exp(get_exponent(x, y, sigma))
            )

    return kernel_x, kernel_y
```

```python
kernel_x, kernel_y = get_derivative_of_gaussian_kernel(5, 0.6)
```

```python
plt.subplot(1, 2, 1)
plt.imshow(kernel_x, cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(kernel_y, cmap="gray")
```

#### Larger Version

```python
kernel_x_large, kernel_y_large = get_derivative_of_gaussian_kernel(51, 25.)

plt.subplot(1, 2, 1)
plt.imshow(kernel_x_large, cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(kernel_y_large, cmap="gray")
```

```python
plt.imshow(edges, cmap="gray")
```

## Distance transform implementation

This code very closely resembles the algorithm in the paper.
http://people.cs.uchicago.edu/~pff/papers/dt.pdf

```python
def find_horizontal_intersection(f, v, q, k):
    """
    f: the function
    v: the locations of the parapolas in lower envelope
    q: the current parabola that we are considering.
    k: number of parabolas in the lower envelope.
    """
    return ((f[q] + q * q) - (f[v[k]] + v[k] * v[k])) / (2 * q - 2 * v[k])


def l2_distance_transform_1d(f, pos_inf, neg_inf):
    d = np.zeros_like(f)  # result
    v = np.zeros_like(f, dtype=np.int32)  # locations of parabolas in lower envelope
    z = np.zeros_like(f)  # locations of boundaries between parabolas
    
    z[0] = neg_inf
    z[1] = pos_inf
    
    # compute the lower envelope
    k = 0
    for q in range(1, len(f)):
        s = find_horizontal_intersection(f, v, q, k)
        while s <= z[k]:
            k -= 1
            s = find_horizontal_intersection(f, v, q, k)
        k += 1
        v[k] = q
        z[k] = s
        z[k + 1] = pos_inf
    
    # fill in the values of distance transform
    k = 0
    for q in range(len(f)):
        while z[k + 1] < q:
            k += 1
        d[q] = np.power(q - v[k], 2) + f[v[k]]
    return d


def l2_distance_transform_2d(edge_function, positive_inf, negative_inf):
    # all non-edge 
    edge_function *= positive_inf
    
    d = np.zeros_like(edge_function)
    for i in range(edge_function.shape[1]):
        d[:, i] = l2_distance_transform_1d(
            edge_function[:, i], positive_inf, negative_inf
        )
    
    edge_function = d
    d = np.zeros_like(edge_function)
    for i in range(edge_function.shape[0]):
        d[i, :] = l2_distance_transform_1d(
            edge_function[i, :], positive_inf, negative_inf
        )
    
    d = np.sqrt(d)
    return d
```

```python
dt_result = l2_distance_transform_2d(
    edges.copy(),
    positive_inf=1024 * 1024 * 2,
    negative_inf=-1024 * 1024 * 2
)
```

## Distance transform using `cv2.distanceTransform`

Let's look at the OpenCV implementation. https://docs.opencv.org/2.4.13.7/modules/imgproc/doc/miscellaneous_transformations.html#distancetransform

> Calculates the distance to the **closest zero pixel** for each pixel of the source image.

> src – 8-bit, single-channel (binary) source image.

> dst – Output image with calculated distances. It is a 32-bit floating-point, single-channel image of the same size as src .

```python
dt_result_cv = cv2.distanceTransform(
    src=np.uint8(edges * 255.0),
    distanceType=cv2.DIST_L2,
    maskSize=cv2.DIST_MASK_PRECISE,
)
```
